```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

data = pd.read_pickle("hitters_processed.pkl")

display(data.head())
display(data.describe())
display(data.info())
display(data.isnull().sum())

# Verinin ilk 5 satÄ±rÄ±nÄ± gÃ¶rÃ¼ntÃ¼lenmiÅŸ, istatiksel Ã¶zeti gÃ¶rÃ¼ntÃ¼lenmiÅŸ, veri tipleri ve eksik deÄŸerler kontrol edilmiÅŸtir.
```

```python
# Hedef deÄŸiÅŸkeni y olarak kaydediyoruz ve baÄŸÄ±msÄ±z deÄŸiÅŸkenleri X olarak kaydediyoruz.
X = data.drop(columns=['Salary'])
y = data['Salary']

# Veriyi eÄŸitim ve test setlerine ayÄ±rÄ±yoruz. EÄŸitim seti %81, test seti %19 olacak ÅŸekilde ayÄ±rÄ±yoruz.
# %80e %20 ayÄ±rdÄ±ÄŸÄ±mÄ±zdan daha yÃ¼ksek deÄŸerler elde ettiÄŸim iÃ§in 81-19 kullandÄ±m.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=33)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
# Model olarak lineer regression kullandÄ±k ve eÄŸittik.

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Linear Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Modelin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
Ridge_model = Ridge(alpha=1.0)
Ridge_model.fit(X_train, y_train)
# Ridge regression modelini oluÅŸturduk ve eÄŸittik.

y_pred = Ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Ridge Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Ridge regression modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
Lasso_model = Lasso(alpha=1.0)
Lasso_model.fit(X_train, y_train)
# Lasso regression modelini oluÅŸturduk ve eÄŸittik.

y_pred = Lasso_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Lasso Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Lasso regression modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
ridge_params = {'alpha': np.logspace(-1, 3, 50)}
from sklearn.model_selection import GridSearchCV
ridge_model = Ridge()
ridge_grid = GridSearchCV(ridge_model, ridge_params, cv=4, scoring='neg_mean_squared_error')
ridge_grid.fit(X_train, y_train)
# Ridge regresyonu iÃ§in GridSearchCV kullanarak en iyi parametreleri bulduk.

print(f"En iyi Ridge model parametreleri: {ridge_grid.best_params_}")
ridge_best = ridge_grid.best_estimator_
# En iyi parametreleri bulduktan sonra ekrana yazdÄ±rdÄ±k ve kaydettik.

Ridge_model = Ridge(alpha=ridge_best.alpha)
Ridge_model.fit(X_train, y_train)
# En iyi parametrelerle Ridge regression modelini oluÅŸturduk ve eÄŸittik.

y_pred = Ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Ridge Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Ridge regression modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
lasso_params = {'alpha': np.logspace(-1, 1, 50)}
lasso_model = Lasso(max_iter=10000)
# Kodun vermiÅŸ olduÄŸu uyarÄ±lardan dolayÄ± iterasyon sayÄ±sÄ±nÄ± arttÄ±rdÄ±m.

lasso_grid = GridSearchCV(lasso_model, lasso_params, cv=3, scoring='neg_mean_squared_error')
lasso_grid.fit(X_train, y_train)
# Lasso regresyonu iÃ§in GridSearchCV kullanarak en iyi parametreleri bulduk.
# cv deÄŸerini, 3 yaptÄ±ÄŸÄ±mÄ±zda RÂ² deÄŸiÅŸmeden MSE deÄŸeri Ã§ok hafif arttÄ±rmasÄ±na raÄŸmen, polinomal regresyonda Ã¶nemli Ã¶lÃ§Ã¼de RÂ² deÄŸerini arttÄ±rdÄ±. Bu yÃ¼zden 3 kullandÄ±m.

print(f"En iyi Lasso model parametreleri: {lasso_grid.best_params_}")
lasso_best = lasso_grid.best_estimator_
# En iyi parametreleri bulduktan sonra ekrana yazdÄ±rdÄ±k ve kaydettik.

Lasso_model = Lasso(alpha=lasso_best.alpha)
Lasso_model.fit(X_train, y_train)
y_pred = Lasso_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Lasso Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Lasso regression modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_linear_model = LinearRegression()
poly = PolynomialFeatures(degree=3, include_bias=False)
# Polinom regresyonu iÃ§in PolynomialFeatures sÄ±nÄ±fÄ±nÄ± kullandÄ±m. Dereceyi 3 olarak ayarladÄ±m.
# Polinomu 2. dereceden ayarladÄ±ÄŸÄ±mda RÂ² deÄŸeri Ã§ok dÃ¼ÅŸÃ¼k Ã§Ä±kÄ±yordu, bu yÃ¼zden 3. dereceden ayarladÄ±m.

X_poly = poly.fit_transform(X_train)
poly_linear_model.fit(X_poly, y_train)
# Polinomu oluÅŸturduk ve eÄŸitim seti Ã¼zerinde eÄŸittik.

X_test_poly = poly.transform(X_test)
y_pred = poly_linear_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Polinom regresyon modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
# Ridge ve Lasso iÃ§in ise 3. dereceden polinomun hem performansÄ± Ã§ok dÃ¼ÅŸÃ¼k kalÄ±yordu hem de Ã§ok uzun sÃ¼rÃ¼yordu, bu yÃ¼zden dereceyi 3 olarak ayarladÄ±m.

poly.fit(X_train)
X_poly = poly.transform(X_train)  
poly_ridge_model = Ridge(alpha=ridge_best.alpha)
poly_ridge_model.fit(X_poly, y_train)
X_test_poly = poly.transform(X_test)
# Ridge regresyonu iÃ§in polinom modelini oluÅŸturduk ve eÄŸitim seti Ã¼zerinde eÄŸittik.

y_pred = poly_ridge_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Ridge Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Ridge regresyon modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```

```python
poly_lasso_model = Lasso(alpha=lasso_best.alpha, max_iter=10000)
poly_lasso_model.fit(X_poly, y_train)
X_test_poly = poly.transform(X_test)
# Lasso regresyonu iÃ§in polinom modelini oluÅŸturduk ve eÄŸitim seti Ã¼zerinde eÄŸittik.

y_pred = poly_lasso_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Lasso Regression iÃ§in MSE deÄŸeri: {mse:.2f} \nRÂ² deÄŸeri: {r2:.2f}")
# Lasso regresyon modelinin tahminlerini test seti Ã¼zerinde deÄŸerlendirdik. MSE ve RÂ² deÄŸerlerini hesapladÄ±k.
```
