```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')

grafik = sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='Set1',legend=True)
grafik._legend.set_title("Sigara en")
plt.title("Toplam hesap ve bahi arasndaki iliki")
plt.xlabel("Toplam hesap")
plt.ylabel("Bahi")
plt.show()

# Toplam hesap ve bahi arasndaki ilikiyi gsteren bir grafik oluturulmutur.
```

Grafikten elde edinilen bilgiye gre toplam hesap arttka verilen bahi de artmaktadr. Ancak sigara ien ve imeyen arasnda gzle grlebilir bir iliki bulunmamaktadr.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Gn ve Toplam hesap arasndaki iliki")
plt.xlabel("Gn")
plt.ylabel("Toplam hesap")
plt.show()

for i in tips["day"].unique():
    med = tips[tips["day"] == i]["total_bill"].median()
    Q1 = tips[tips["day"] == i]["total_bill"].quantile(0.25)
    Q3 = tips[tips["day"] == i]["total_bill"].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    print(f"{i} gn iin medyan: {med}, alt snr: {alt_sinir:.2f}, st snr: {ust_sinir:.2f}, IQR: {IQR:.2f}")

# Gn ve Toplam hesap arasndaki ilikiyi gsteren bir boxplot grafii oluturulmutur.
# Her gn iin medyan, alt snr, st snr ve IQR deerleri hesaplanmtr.
```

Grafikten elde ettiimiz bilgiye gre haftasonu (Cumartesi, Pazar) toplam hesapta art grlmektedir.

```python
tips["tip_rate"] = tips["tip"] / tips["total_bill"] * 100
sns.barplot(x='size', y='tip_rate', data=tips)
plt.title("Masa bykl ve bahi oran arasndaki iliki")
plt.xlabel("Masa bykl")
plt.ylabel("Bahi oran")
plt.show()

# Masa bykl ve bahi oran arasndaki ilikiyi gsteren bir barplot grafii oluturulmutur.
```

Grafikten elde ettiimiz bilgiye gre 1. tip masa bykl en yksek bahi oranna sahiptir.

```python
sns.histplot(data=tips, x='tip', hue='time', kde=True, palette='Set1')
plt.title("Bahi miktar ve zaman arasndaki iliki")
plt.xlabel("Bahi miktar")
plt.ylabel("Frekans")
plt.legend(title="Zaman", labels=['Lunch', 'Dinner'])
plt.show()

# Bahi miktar ve zaman arasndaki ilikiyi gsteren bir histogram grafii oluturulmutur.
```

Grafiin bize verdii bilgiye gre akam yemei (Dinner) sresinde verilen bahi miktar artmaktadr.

```python
nums = tips.select_dtypes(include=['number'])
corr = nums.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrisi')
plt.show()

# Deikenler arasndaki ilikiyi gsteren korelasyon matrisi oluturulmutur.
```

Grafikten elde ettiimiz bilgiye gre hesap miktar ve bahi ikilisi ile hesap miktar ve masa bykl arasnda yksek korelasyon bulunmaktadr.
