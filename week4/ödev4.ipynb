```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')

grafik = sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='Set1',legend=True)
grafik._legend.set_title("Sigara İçen")
plt.title("Toplam hesap ve bahşiş arasındaki ilişki")
plt.xlabel("Toplam hesap")
plt.ylabel("Bahşiş")
plt.show()

# Toplam hesap ve bahşiş arasındaki ilişkiyi gösteren bir grafik oluşturulmuştur.
```

Grafikten elde edinilen bilgiye göre toplam hesap arttıkça verilen bahşiş de artmaktadır. Ancak sigara içen ve içmeyen arasında gözle görülebilir bir ilişki bulunmamaktadır.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Gün ve Toplam hesap arasındaki ilişki")
plt.xlabel("Gün")
plt.ylabel("Toplam hesap")
plt.show()

for i in tips["day"].unique():
    med = tips[tips["day"] == i]["total_bill"].median()
    Q1 = tips[tips["day"] == i]["total_bill"].quantile(0.25)
    Q3 = tips[tips["day"] == i]["total_bill"].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    print(f"{i} günü için medyan: {med}, alt sınır: {alt_sinir:.2f}, üst sınır: {ust_sinir:.2f}, IQR: {IQR:.2f}")

# Gün ve Toplam hesap arasındaki ilişkiyi gösteren bir boxplot grafiği oluşturulmuştur.
# Her gün için medyan, alt sınır, üst sınır ve IQR değerleri hesaplanmıştır.
```

Grafikten elde ettiğimiz bilgiye göre haftasonu (Cumartesi, Pazar) toplam hesapta artış görülmektedir.

```python
tips["tip_rate"] = tips["tip"] / tips["total_bill"] * 100
sns.barplot(x='size', y='tip_rate', data=tips)
plt.title("Masa büyüklüğü ve bahşiş oranı arasındaki ilişki")
plt.xlabel("Masa büyüklüğü")
plt.ylabel("Bahşiş oranı")
plt.show()

# Masa büyüklüğü ve bahşiş oranı arasındaki ilişkiyi gösteren bir barplot grafiği oluşturulmuştur.
```

Grafikten elde ettiğimiz bilgiye göre 1. tip masa büyüklüğü en yüksek bahşiş oranına sahiptir.

```python
sns.histplot(data=tips, x='tip', hue='time', kde=True, palette='Set1')
plt.title("Bahşiş miktarı ve zaman arasındaki ilişki")
plt.xlabel("Bahşiş miktarı")
plt.ylabel("Frekans")
plt.legend(title="Zaman", labels=['Lunch', 'Dinner'])
plt.show()

# Bahşiş miktarı ve zaman arasındaki ilişkiyi gösteren bir histogram grafiği oluşturulmuştur.
```

Grafiğin bize verdiği bilgiye göre akşam yemeği (Dinner) süresinde verilen bahşiş miktarı artmaktadır.

```python
nums = tips.select_dtypes(include=['number'])
corr = nums.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrisi')
plt.show()

# Değişkenler arasındaki ilişkiyi gösteren korelasyon matrisi oluşturulmuştur.
```

Grafikten elde ettiğimiz bilgiye göre hesap miktarı ve bahşiş ikilisi ile hesap miktarı ve masa büyüklüğü arasında yüksek korelasyon bulunmaktadır.
