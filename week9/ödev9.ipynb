```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

data = pd.read_pickle("hitters_processed.pkl")

display(data.head())
display(data.describe())
display(data.info())
display(data.isnull().sum())

# Verinin ilk 5 satırını görüntülenmiş, istatiksel özeti görüntülenmiş, veri tipleri ve eksik değerler kontrol edilmiştir.
```

```python
# Hedef değişkeni y olarak kaydediyoruz ve bağımsız değişkenleri X olarak kaydediyoruz.
X = data.drop(columns=['Salary'])
y = data['Salary']

# Veriyi eğitim ve test setlerine ayırıyoruz. Eğitim seti %81, test seti %19 olacak şekilde ayırıyoruz.
# %80e %20 ayırdığımızdan daha yüksek değerler elde ettiğim için 81-19 kullandım.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=33)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
# Model olarak lineer regression kullandık ve eğittik.

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Linear Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Modelin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
Ridge_model = Ridge(alpha=1.0)
Ridge_model.fit(X_train, y_train)
# Ridge regression modelini oluşturduk ve eğittik.

y_pred = Ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Ridge Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Ridge regression modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
Lasso_model = Lasso(alpha=1.0)
Lasso_model.fit(X_train, y_train)
# Lasso regression modelini oluşturduk ve eğittik.

y_pred = Lasso_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Lasso Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Lasso regression modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
ridge_params = {'alpha': np.logspace(-1, 3, 50)}
from sklearn.model_selection import GridSearchCV
ridge_model = Ridge()
ridge_grid = GridSearchCV(ridge_model, ridge_params, cv=4, scoring='neg_mean_squared_error')
ridge_grid.fit(X_train, y_train)
# Ridge regresyonu için GridSearchCV kullanarak en iyi parametreleri bulduk.

print(f"En iyi Ridge model parametreleri: {ridge_grid.best_params_}")
ridge_best = ridge_grid.best_estimator_
# En iyi parametreleri bulduktan sonra ekrana yazdırdık ve kaydettik.

Ridge_model = Ridge(alpha=ridge_best.alpha)
Ridge_model.fit(X_train, y_train)
# En iyi parametrelerle Ridge regression modelini oluşturduk ve eğittik.

y_pred = Ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Ridge Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Ridge regression modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
lasso_params = {'alpha': np.logspace(-1, 1, 50)}
lasso_model = Lasso(max_iter=10000)
# Kodun vermiş olduğu uyarılardan dolayı iterasyon sayısını arttırdım.

lasso_grid = GridSearchCV(lasso_model, lasso_params, cv=3, scoring='neg_mean_squared_error')
lasso_grid.fit(X_train, y_train)
# Lasso regresyonu için GridSearchCV kullanarak en iyi parametreleri bulduk.
# cv değerini, 3 yaptığımızda R² değişmeden MSE değeri çok hafif arttırmasına rağmen, polinomal regresyonda önemli ölçüde R² değerini arttırdı. Bu yüzden 3 kullandım.

print(f"En iyi Lasso model parametreleri: {lasso_grid.best_params_}")
lasso_best = lasso_grid.best_estimator_
# En iyi parametreleri bulduktan sonra ekrana yazdırdık ve kaydettik.

Lasso_model = Lasso(alpha=lasso_best.alpha)
Lasso_model.fit(X_train, y_train)
y_pred = Lasso_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Lasso Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Lasso regression modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_linear_model = LinearRegression()
poly = PolynomialFeatures(degree=3, include_bias=False)
# Polinom regresyonu için PolynomialFeatures sınıfını kullandım. Dereceyi 3 olarak ayarladım.
# Polinomu 2. dereceden ayarladığımda R² değeri çok düşük çıkıyordu, bu yüzden 3. dereceden ayarladım.

X_poly = poly.fit_transform(X_train)
poly_linear_model.fit(X_poly, y_train)
# Polinomu oluşturduk ve eğitim seti üzerinde eğittik.

X_test_poly = poly.transform(X_test)
y_pred = poly_linear_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Polinom regresyon modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
# Ridge ve Lasso için ise 3. dereceden polinomun hem performansı çok düşük kalıyordu hem de çok uzun sürüyordu, bu yüzden dereceyi 3 olarak ayarladım.

poly.fit(X_train)
X_poly = poly.transform(X_train)  
poly_ridge_model = Ridge(alpha=ridge_best.alpha)
poly_ridge_model.fit(X_poly, y_train)
X_test_poly = poly.transform(X_test)
# Ridge regresyonu için polinom modelini oluşturduk ve eğitim seti üzerinde eğittik.

y_pred = poly_ridge_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Ridge Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Ridge regresyon modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

```python
poly_lasso_model = Lasso(alpha=lasso_best.alpha, max_iter=10000)
poly_lasso_model.fit(X_poly, y_train)
X_test_poly = poly.transform(X_test)
# Lasso regresyonu için polinom modelini oluşturduk ve eğitim seti üzerinde eğittik.

y_pred = poly_lasso_model.predict(X_test_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Polynomial Lasso Regression için MSE değeri: {mse:.2f} \nR² değeri: {r2:.2f}")
# Lasso regresyon modelinin tahminlerini test seti üzerinde değerlendirdik. MSE ve R² değerlerini hesapladık.
```

Yapmış olduğumuz testler sonucunda maalesef polinomal regresyon, modelin performansını düşürdü. Yapmış olduğumuz testlere baktığımızda görüyoruz ki, alphayı 1 olarak ayarladığımız lasso ve ridge modelleri en başarılı
sonuçları vermişlerdir. En iyi alphayı arama methodu ile belirlediğimizde performansın düşme sebebi cv değerinin düşük ayarlanmasından kaynaklıdır. Polinomal regresyon modellerinin başarısını arttırdığından dolayı
cv değerlerini düşük seçtim.
